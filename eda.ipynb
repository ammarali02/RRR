```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import data_manipulation as data

%matplotlib inline
```

```python
data.random_sample("feature-data/", 'sample-data/', 0.05)
```

```python
df = data.to_dataframe('sample-data')
df
```

```python
df_all = df.groupby('Category').mean()
df_all
```

```python
count_df = df.groupby('Category').size().reset_index(name='Count')

df_all = pd.merge(df_all, count_df, on = 'Category')
df_all
```

```python
plt.figure(figsize=(15,4))
plot = sns.barplot(data = df_all, x = 'Category', y = 'Count')
plot.set_xticklabels(plot.get_xticklabels(), rotation = 40, fontsize=7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,4))
plot = sns.barplot(data = df_all, x = 'Category', y = 'Width')
plot.set_xticklabels(plot.get_xticklabels(), rotation = 40, fontsize=7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,4))
plot = sns.barplot(data = df_all, x = 'Category', y = 'Height')
plot.set_xticklabels(plot.get_xticklabels(), rotation = 40, fontsize=7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,4))
plot = sns.barplot(data = df_all, x = 'Category', y = 'Mean')
plot.set_xticklabels(plot.get_xticklabels(), rotation = 40, fontsize=7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,4))
plot = sns.barplot(data = df_all, x = 'Category', y = 'Standard Deviation')
plot.set_xticklabels(plot.get_xticklabels(), rotation = 40, fontsize=7)
plt.tight_layout()
plt.show()
```

```python
data.random_sample("feature-data", 'sample-data-large/', 0.25)
df_large = data.to_dataframe('sample-data-large/')
df_large.reset_index(inplace=True)
```

```python
g = sns.relplot(data=df_large, x='Width', y='Height',col='Category', col_wrap=4, height=4, aspect=1.5, kind='scatter', legend = False)
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(40)
        label.set_fontsize(7)

g.figure.suptitle('Aspect Ratio per Category', fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.95)
plt.show()
```
