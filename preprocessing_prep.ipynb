```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import data_manipulation as data

%matplotlib inline
```

```python
df = data.to_dataframe('feature-data')
```

Appears that 500 is a good value to have as a minimum per category. Need to check whether the following categories are worth keeping: "other metal objects", "news paper", "smartphones", "gloves". Could also use data augmentation to increase counts for those if necessary.

```python
count_df = df.groupby('Category').size().reset_index(name='Count')
count_df.sort_values(by=['Count'])
```

Checking and working with Null Data. There exists 5 null values in the follwoing indexes. Looks safe to just go ahead and drop the values. 

```python
df_dropped = df.dropna(inplace=False, how='any')
count_df_dropped = df_dropped.groupby('Category').size().reset_index(name='Count')
count_df_dropped.sort_values(by=['Count'])

count_df['Count'].subtract(count_df_dropped['Count'])
```
