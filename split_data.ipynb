```python
"""
This code takes a random sample from the feature-data folder and splits the images into test, train, and validate
subfolders based on specified params
"""


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import data_manipulation as data
import shutil
import os

%matplotlib inline
```

```python
random_sample_percent = 0.15
```

```python
data.random_sample("feature-data/", random_sample_percent)
```

Note: the code below should take about 30 seconds to run! (you're loading 2500+ images)

```python
df = data.to_dataframe('sample-data')
df
```

```python
# splits into 80/10/10 split
train, validate, test = np.split(df.sample(frac=1, random_state=42), 
                       [int(.8*len(df)), int(.9*len(df))])
```

```python
for indices, row in train.iterrows():
	print(indices)
```

```python
# Create the main directory
main_dir = 'split-data'
os.makedirs(main_dir, exist_ok=True)
folder_to_df = {"train": train, "test": test, "validate": validate}

# Loop through each row in the DataFrame

for folder_name, df in folder_to_df.items():
    sub_main_dir = os.path.join(main_dir, folder_name)
    os.makedirs(sub_main_dir, exist_ok=True)
    for indices, row in df.iterrows():
        category = indices[0]
        image_name = indices[1]
        
        # Create subfolder for each category
        category_dir = os.path.join(sub_main_dir, category)
        os.makedirs(category_dir, exist_ok=True)
        
        # Path to the source image file (assuming they are in the current directory)
        src_path = fr"./sample-data/{category}/{image_name}"
        
        # Path to the destination
        dst_path = os.path.join(category_dir, image_name)
        
        # Copy the image file to the destination
        shutil.copy(src_path, dst_path)
```
